```python
df = pd.read_excel('성별 첫일자리 그만둔 사유.xlsx')
```

```python
# 데이터 상위 5개 행
df.head()

# 데이터 구조 확인
df.info()

# 통계 요약
df.describe()
```

```python
df = df[df['연령구분'] != '20~34세'] #20~34세 행은 제거
df = df.drop(index = [5,6]) #나머지 불필요 행도 제거
df.drop(df.columns[:2], axis=1, inplace=True) #0,1번째 열 제거
df.index = ['퇴직사유', '전체', '남자', '여자'] #행 이름 설정
df
```

```python
df.head()
# 0번째 행을 인덱스로 지정
df.columns = df.iloc[0]      # 0번째 행을 열 이름으로
df = df[1:]                  # 0번째 행 제거
df = df.drop(df.columns[0], axis=1)
df
```

```python
#맨 앞의 열을 인덱스로 지정
df.set_index(df.columns[0], inplace=True)
df.index.name = '퇴직사유'
df
```
